### What is the most 10 popular song?

```python
most_popular_song=df.query('popularity>90', inplace=False).sort_values('popularity', ascending= False)
most_popular_song[:10]
```

```python
#changing release_date index
df.set_index('release_date', inplace= True)
df.index=pd.to_datetime(df.index)
```

### Converting Duration_ms with duration

```python
df['duration']=df['duration_ms'].apply(lambda x : round(x/1000))
df.drop('duration_ms', inplace=True, axis=1)
```

```python
df.duration.head()
```

### Creating First Visualization with Correlation with Using Pearson Method

```python
corr_df=df.drop(['key','mode','explicit'], axis=1).corr(method='pearson')
plt.figure(figsize=(14,8))
heatmap=sns.heatmap(corr_df, annot=True, fmt='.1g')
heatmap.set_title('Correlation Heatmap Between Variable')
heatmap.set_xticklabels(heatmap.get_xticklabels(), rotation=90)
```

```python
smp_df=df.sample(int(0.004*len(df)))
```

```python
print(len(smp_df))
```

#### Loudness Vs Energy

```python
plt.figure(figsize=(10,6))
sns.regplot(data=smp_df, y='loudness', x='energy', color='r').set(title='Loudness Vs Energy')
```

#### Popularity Vs Acousticness

```python
plt.figure(figsize=(10,6))
sns.regplot(data=smp_df, y='popularity', x='acousticness', color='c').set(title='Popularity Vs Acousticness')
```

#### Speechiness Vs Liveness

```python
plt.figure(figsize=(10,6))
sns.regplot(data=smp_df, y='speechiness', x='liveness', color='b').set(title='Speechiness Vs Liveness')
```

### 

```python
df['dates']=df.index.get_level_values('release_date')
df.dates=pd.to_datetime(df.dates)
```

```python
years=df.dates.dt.year
```

```python
sns.displot(years, discrete=True , aspect=2, height=5 , kind='hist').set(title='Number Of Song Per Years')
```

#### Year Vs Duration

```python
total_dr=df.duration
```

```python
fig_dims=(20,8)
fig, ax =plt.subplots(figsize= fig_dims)
fig=sns.barplot(x=years,y=total_dr, ax=ax,errwidth=False).set(title='Year Vs Duration')
plt.xticks(rotation=90)
```

```python
total_dr=df.duration
sns.set_style(style='whitegrid')
fig_dims=(10,5)
fig, ax =plt.subplots(figsize= fig_dims)
fig=sns.lineplot(x=years,y=total_dr, ax=ax).set(title='Year Vs Duration')
plt.xticks(rotation=60)
```
